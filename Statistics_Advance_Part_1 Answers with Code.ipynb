5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

 A Cumulative Distribution Function (CDF) provides the probability that a random variable is less than or equal to a certain value.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x, 0, 1)
cdf = norm.cdf(x, 0, 1)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, label="PDF", color='blue')
plt.plot(x, cdf, label="CDF", color='red')
plt.legend()
plt.title("Probability Density Function (PDF) vs Cumulative Distribution Function (CDF)")
plt.show()
```

7.  What are the key properties of a Bernoulli distribution?

 A Bernoulli Distribution represents a binary outcome (e.g., success/failure). It is defined using a probability p.




```python
from scipy.stats import bernoulli

p = 0.5
data = bernoulli.rvs(p, size=100)

plt.hist(data, bins=[-0.5, 0.5, 1.5], density=True, edgecolor='black')
plt.xticks([0, 1], ["Failure", "Success"])
plt.title("Bernoulli Distribution (p=0.5)")
plt.show()
```

16.  How do you calculate a Z-score, and what does it represent.

 A Z-score measures how far a value is from the mean in standard deviations.



```python
import scipy.stats as stats

def z_score(x, mu, sigma):
    return (x - mu) / sigma

x = 100
mu = 90
sigma = 10
print("Z-score:", z_score(x, mu, sigma))
```
